```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import collections
```

```python
for i in range(1,46):

    try:
        y = str(i).zfill(6)
        print(y)
        # read the image
        image = cv2.imread("./demo-frames/raft_flo_rgb/mask_{}.png".format(y))

        # convert to RGB
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # reshape the image to a 2D array of pixels and 3 color values (RGB)
        pixel_values = image.reshape((-1, 3))
        # convert to float
        pixel_values = np.float32(pixel_values)

        print(pixel_values.shape)

        # define stopping criteria
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)


        # number of clusters (K)
        k = 8
        _, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_PP_CENTERS)
        # print(labels)

        # convert back to 8 bit values
        centers = np.uint8(centers)

        # flatten the labels array
        labels = labels.flatten()

        # convert all pixels to the color of the centroids
        # 畫出分完群的樣子
        kmeans_img = centers[labels.flatten()]
        kmeans_img = kmeans_img.reshape(image.shape)
        # show the image
        plt.imshow(kmeans_img)
        plt.show()

        
        # 找出每個群的數量
        dic = collections.Counter(labels.flatten())
        dic = {k: v for k, v in sorted(dic.items(), key=lambda item: item[1])}
        print(dic)
        #######################################################
        sort_groupid = [gid for gid in dic]
        split = 0
        foreground_gid = sort_groupid[0]
        background_gid = 0


        for i in range(len(sort_groupid)-1):
            if dic[sort_groupid[i]] < dic[sort_groupid[i+1]]/4:
                background_gid = sort_groupid[i+1]
                split = i+1

        # 前景
        for i in range(split):
            labels[labels==sort_groupid[i]] = foreground_gid

        # 背景
        for i in range(split,len(sort_groupid)):
            labels[labels==sort_groupid[i]] = background_gid

        ########################################################

        # convert all pixels to the color of the centroids
        segmented_image = centers[labels.flatten()]


        # reshape back to the original image dimension
        segmented_image = segmented_image.reshape(image.shape)
        
        # 將背景轉成黑色，人物轉成白色，製造出mask
        background_rgb = segmented_image[0][400]
        new_segmented_image = np.where(segmented_image == background_rgb, [0,0,0], segmented_image)
        mask = np.where(new_segmented_image != [0,0,0], [255,255,255], new_segmented_image)

        cv2.imwrite("./demo-frames/raft_kmeans_mask/mask_{}.png".format(y), mask)


        
    except:
        print("沒有這個raft_rgb影像 ", y)
```
